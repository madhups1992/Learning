# Inputs:

```python
class Solution(object):
    def divide(self, dividend, divisor):
        """
        :type dividend: int
        :type divisor: int
        :rtype: int
        """
        val=0
        neg=1
        if(dividend<0):
            neg*=-1
            dividend*=-1
        if(divisor<0):
            neg*=-1
            divisor*=-1
        while dividend>divisor:
            dividend = dividend-divisor
            val+=1
        return(val*neg)
```

```python
S = Solution()
S.divide(7,-3)
```
