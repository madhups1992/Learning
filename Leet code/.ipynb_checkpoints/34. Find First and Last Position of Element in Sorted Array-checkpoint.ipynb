```python
class Solution(object):
    def first_occ(self,nums, target):
        l,r = 0,len(nums)-1
        while(l<=r):
            mid = (r+l) //2

            if(nums[mid]<target):
                l=mid+1
            else:
                r=mid-1
        return(l)
    def last_occ(self,nums, target):
        l,r = 0,len(nums)-1
        while(l<=r):
            mid = (r+l) //2

            if(nums[mid]>target):
                r=mid-1
            else:
                l=mid+1
        return(r)

        
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        first_ind =-1
        last_ind = -1
        l,r = 0,len(nums)-1
        if(len(nums)==0): 
            return([-1,-1])
        first_ind = self.first_occ( nums, target)
        last_ind = self.last_occ( nums, target)
        
        return(first_ind,last_ind) if first_ind<=last_ind else [-1,-1]

            
        
```
