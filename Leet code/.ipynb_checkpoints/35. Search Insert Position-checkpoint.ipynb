https://leetcode.com/problems/search-insert-position/

```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        l,r=0,len(nums)-1
        
        while(l<=r):
            mid = (l+r) // 2
            if(target <= nums[mid]):
                r=mid-1
            else:
                l=mid+1
        return(l)
```

```python
s= Solution()
s.searchInsert(nums = [1,3,5,6], target = 7)
```
