```python
class Solution(object):
    def countAndSay(self, n):
        """
        :type n: int
        :rtype: str
        """
        s=''
        temp ='1'
        for n_i in range(n-1):
            s= temp
            temp=''
            count=1
            for i in range(len(s)-1):
                if(s[i]==s[i+1]):
                    count+=1
                else:
                    temp+=str(count)+s[i]
                    count=1
            temp+=str(count)+s[len(s)-1]
            print(temp)
        return(temp)
            
            
        
```

```python
s = Solution()
s.countAndSay(6)
```
