```python
class Solution(object):
    def multiply(self, num1, num2):
        """
        :type num1: str
        :type num2: str
        :rtype: str
        """
        result=[0]*(len(num1)+len(num2))
        num1=num1[::-1]
        num2=num2[::-1]
        
        for i1 in range(len(num1)):
            for i2 in range(len(num2)):
                digit = int(num1[i1])*int(num2[i2])
                result[i1+i2] += digit % 10
                result[i1+i2+1] += (digit // 10) + (result[i1+i2]//10)
                result[i1+i2]%=10
                
        result = result[::-1]
        
        init=0
        while(init <len(result) and result[init]==0):
            init+=1
        
        return(''.join(map(str,result[init:])))
        
        
            
```

```python
s= Solution()
s.multiply(num1='624',num2='35')
# 426 53
```
